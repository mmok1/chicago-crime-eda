Retrieve the type of crime from the crime table, and retrieve the temperature from the weather table;
Combine the crime table with the weather table, and ensure that records are matched based on the date;
Group the results by crime type, each crime type appears only once with a corresponding temperature

```python
#checking the type of crime type and temprature
query = """select cr.primary_type, we.temp
from crime as cr
inner join weather as we on
DATE(cr.date_only) = DATE(we.datetime)
GROUP BY primary_type;
"""

final = pd.read_sql(query, conn)
final
```

Categorize temperatures into five weather categories and counts the number of crimes that occurred in each category by joining the crime and weather tables based on matching dates.
Crimes tend to happen on warm days.

```python
#crime counts based on temp
query ="""SELECT 
    CASE 
        WHEN we.temp < 32 THEN 'Below Freezing (<32°F)'
        WHEN we.temp BETWEEN 32 AND 50 THEN 'Cold (32-50°F)'
        WHEN we.temp BETWEEN 51 AND 70 THEN 'Mild (51-70°F)'
        WHEN we.temp BETWEEN 71 AND 90 THEN 'Warm (71-90°F)'
        ELSE 'Hot (>90°F)'
    END AS temp_category,
    COUNT(cr.id) AS crime_count
FROM crime cr
INNER JOIN weather we ON DATE(cr.date_only) = DATE(we.datetime) 
GROUP BY temp_category
ORDER BY temp_category ASC;

"""
final = pd.read_sql(query, conn)
final
```

Categorize daylight hours into different groups, count the number of crimes in each category, and sort them in descending order of crime count.
There are more crimes when daylight hours are between 9 to 12.

```python
#crime and daylight hours
query ="""SELECT 
    CASE 
        WHEN we.daylight_hours < 6 THEN 'Very Short (<6 hrs)'
        WHEN we.daylight_hours BETWEEN 6 AND 9 THEN 'Short (6-9 hrs)'
        WHEN we.daylight_hours BETWEEN 9 AND 12 THEN 'Medium (9-12 hrs)'
        WHEN we.daylight_hours BETWEEN 12 AND 15 THEN 'Long (12-15 hrs)'
        ELSE 'Very Long (>15 hrs)'
    END AS daylight_category,
    COUNT(cr.id) AS crime_count
FROM crime cr
INNER JOIN weather we ON DATE(cr.date_only) = DATE(we.datetime)
GROUP BY daylight_category
ORDER BY crime_count DESC;
"""
final = pd.read_sql(query, conn)
final
```

Calculate the average visibility for crimes where an arrest was made and where no arrest was made.
It joins the crime and weather data based on matching dates.
However, visibility conditions appear to have a minimal effect on whether an arrest is made.

```python
#arrest and visibility
query = """
SELECT cr.arrest, AVG(we.visibility) AS avg_visibility
FROM crime AS cr
INNER JOIN weather AS we 
ON DATE(cr.date_only) = DATE(we.datetime)
GROUP BY cr.arrest;
"""

final = pd.read_sql(query, conn)
final
```

Identify blocks where the total number of arrests is 300 or more, ranking them in descending order.
The block "001XX N STATE ST" had 338 arrests. So this location may be a crime hotspot or a highly policed area. 
(Based on the data from the Chicago Police Department, we have hidden sensitive information here.)

```python
#crime and num_arrests
query = """
SELECT block,sum(arrest) AS num_arrests 
FROM crime
GROUP BY block 
HAVING sum(arrest) >= 300 
ORDER BY num_arrests DESC;""" 

final = pd.read_sql(query, conn)
final
```

Identify dangerous blocks based on the total number of arrests recorded, categorizing them into differest arrest ranges and sorting them in descending order.
We have a few dangerous and high arrests blocks here.

```python
#dangerous blocks
query = """
SELECT 
    block,
    CASE 
        WHEN total_arrests >= 250 THEN '250~'
        WHEN total_arrests >= 200 THEN '200~'
        WHEN total_arrests >= 150 THEN '150~'
        ELSE '100~'
    END AS num_arrests
FROM (
    SELECT block, SUM(arrest) AS total_arrests
    FROM crime
    GROUP BY block
) AS crime_summary
WHERE total_arrests >= 100
ORDER BY num_arrests DESC, total_arrests DESC;

"""

final = pd.read_sql(query, conn)
final
```

Retrieve crime types that have 100 or fewer total arrests, sorting them in ascending order by the number of arrest
Human trafficking has 0 arrests suggests that it might be underreporting or difficult in catching and prosecuting offenders. We have low arrests in crimes like public indecency, gambling and obscenity, and more arrests in arson, criminal sexual assault, and sex offense.

```python
#primary type and number of arrests
query = """
SELECT primary_type,sum(arrest) AS num_arrests 
FROM crime
GROUP BY primary_type 
HAVING sum(arrest) <= 100 
ORDER BY num_arrests ASC;""" 

final = pd.read_sql(query, conn)
final
```

Analyze weather conditions on days when arrests were made or not made and calculate a running total of visibility over time.

```python
# Arrest and visibility
query = """
SELECT cr.arrest,
       we.visibility,
       we.humidity,
       we.snow,
       we.windspeed,
       we.daylight_hours,
       SUM(we.visibility) OVER (PARTITION BY cr.arrest ORDER BY cr.date_only) AS running_visibility
FROM crime AS cr
INNER JOIN weather AS we
  ON DATE(cr.date_only) = DATE(we.datetime)
ORDER BY cr.date_only;
"""
final = pd.read_sql(query, conn)
final
```

Analyze the relationship between crime types and temperature by calculating the average temperature for each crime type.

"WITH" statement creates a Common Table Expression called ranked_crimes;
"INNER JOIN weather AS we" joins the crime and weather tables on the same date;
"ROW_NUMBER() OVER (PARTITION BY cr.primary_type ORDER BY we.temp) AS temp_rank": we rank temperature values within each crime type and assign a unique rank to each crime's temperature, sorted from lowest to highest.

We calculate the average temperature associated with each crime type and group crimes by primary type, ensuring that each row represents one crime type and its average temperature, sorting crime types in ascending order of average temperature.

```python
#primary type and temprature
query = """
WITH ranked_crimes AS (
    SELECT cr.primary_type, 
           we.temp,
           ROW_NUMBER() OVER (PARTITION BY cr.primary_type ORDER BY we.temp) AS temp_rank
    FROM crime AS cr
    INNER JOIN weather AS we 
        ON DATE(cr.date_only) = DATE(we.datetime)
)
SELECT primary_type, 
       AVG(temp) AS avg_temp
FROM ranked_crimes
GROUP BY primary_type
ORDER BY avg_temp;
"""

# Assuming 'conn' is your database connection
final = pd.read_sql(query, conn)
final
```
